## 4. Batalha 1 — A tirania dos tipos de dados

Nos slides, vimos que datas não podem ser tratadas como texto e que números em formato string quebram análises. fileciteturn2file0

### Tarefa
Converta corretamente, quando fizer sentido:
- `data`
- `receita`
- `lucro`
- `quantidade`

### Orientação
- Investigue valores estranhos antes de converter
- Pense no uso de `errors='coerce'`
- Registre em markdown ou comentário quais problemas encontrou


```python
# ## 4. Batalha 1 — A tirania dos tipos de dados

print("=== BATALHA 1: CONVERSÃO DE TIPOS DE DADOS ===\n")

# 1. Investigar valores estranhos antes de converter

print("1. Investigando coluna 'data':")
print(f"   Tipo atual: {df['data'].dtype}")
print(f"   Valores únicos (primeiros 10): {df['data'].unique()[:10]}")
print(f"   Possíveis problemas: {df['data'].isnull().sum()} nulos")

# Verificar se há datas em formato inconsistente
datas_problematicas = df[~df['data'].str.match(r'^\d{4}-\d{2}-\d{2}$', na=False)]
if len(datas_problematicas) > 0:
    print(f"   ⚠️ Encontradas {len(datas_problematicas)} datas em formato inconsistente")
    print(f"   Exemplos: {datas_problematicas['data'].unique()[:5]}")

print("\n2. Investigando coluna 'receita':")
print(f"   Tipo atual: {df['receita'].dtype}")
# Verificar se há valores que não podem ser convertidos para float
receita_nao_numerica = df[~df['receita'].astype(str).str.replace(',', '.').str.replace('-', '', regex=False).str.replace('.', '', regex=False).str.isdigit()]
if len(receita_nao_numerica) > 0:
    print(f"   ⚠️ Possíveis valores não numéricos: {receita_nao_numerica['receita'].unique()[:10]}")

print("\n3. Investigando coluna 'lucro':")
print(f"   Tipo atual: {df['lucro'].dtype}")

print("\n4. Investigando coluna 'quantidade':")
print(f"   Tipo atual: {df['quantidade'].dtype}")

# Realizar as conversões de tipo

# Copiar DataFrame para trabalhar
df_clean = df.copy()

# 1. Converter coluna 'data' para datetime
try:
    # Tentar converter com errors='coerce' para transformar valores inválidos em NaT
    df_clean['data'] = pd.to_datetime(df_clean['data'], errors='coerce')
    print("✅ Coluna 'data' convertida para datetime")
    print(f"   Datas válidas: {df_clean['data'].notna().sum()}")
    print(f"   Datas inválidas (transformadas em NaT): {df_clean['data'].isna().sum()}")
except Exception as e:
    print(f"❌ Erro na conversão: {e}")

# 2. Converter coluna 'receita' para float
try:
    # Remover possíveis caracteres especiais e converter
    df_clean['receita'] = df_clean['receita'].astype(str).str.replace(',', '.').str.replace('R$', '').str.replace(' ', '')
    df_clean['receita'] = pd.to_numeric(df_clean['receita'], errors='coerce')
    print("✅ Coluna 'receita' convertida para float")
    print(f"   Valores válidos: {df_clean['receita'].notna().sum()}")
    print(f"   Valores inválidos: {df_clean['receita'].isna().sum()}")
except Exception as e:
    print(f"❌ Erro na conversão: {e}")

# 3. Converter coluna 'lucro' para float
try:
    df_clean['lucro'] = df_clean['lucro'].astype(str).str.replace(',', '.').str.replace('R$', '').str.replace(' ', '')
    df_clean['lucro'] = pd.to_numeric(df_clean['lucro'], errors='coerce')
    print("✅ Coluna 'lucro' convertida para float")
    print(f"   Valores válidos: {df_clean['lucro'].notna().sum()}")
    print(f"   Valores inválidos: {df_clean['lucro'].isna().sum()}")
except Exception as e:
    print(f"❌ Erro na conversão: {e}")

# 4. Converter coluna 'quantidade' para int
try:
    df_clean['quantidade'] = pd.to_numeric(df_clean['quantidade'], errors='coerce')
    print("✅ Coluna 'quantidade' convertida para numérica")
    print(f"   Valores válidos: {df_clean['quantidade'].notna().sum()}")
    print(f"   Valores inválidos: {df_clean['quantidade'].isna().sum()}")
except Exception as e:
    print(f"❌ Erro na conversão: {e}")

print("\n=== Tipos após conversão ===")
print(df_clean.dtypes)
```

### Tarefa

```python
# ## 5. Batalha 2 — O enigma dos valores ausentes

print("=== BATALHA 2: TRATAMENTO DE VALORES AUSENTES ===\n")

# 1. Mapear valores ausentes por coluna
print("1. Mapeamento de valores ausentes:")
nulos_antes = df_clean.isnull().sum()
print(nulos_antes[nulos_antes > 0])

# 2. Identificar colunas críticas com nulos
print("\n2. Colunas com valores ausentes:")
for col in df_clean.columns:
    if df_clean[col].isnull().sum() > 0:
        pct = (df_clean[col].isnull().sum() / len(df_clean)) * 100
        print(f"   - {col}: {df_clean[col].isnull().sum()} ausentes ({pct:.2f}%)")

# 3. Analisar padrão de nulos
print("\n3. Análise de padrão de nulos:")
# Verificar se os nulos estão concentrados nas mesmas linhas
linhas_com_nulos = df_clean[df_clean.isnull().any(axis=1)]
print(f"   Linhas que contêm pelo menos um nulo: {len(linhas_com_nulos)} ({len(linhas_com_nulos)/len(df_clean)*100:.2f}%)")

# Verificar se há nulos em colunas críticas
colunas_criticas = ['receita', 'lucro', 'data', 'canal_venda']
print("\n4. Nulos em colunas críticas:")
for col in colunas_criticas:
    if col in df_clean.columns:
        print(f"   - {col}: {df_clean[col].isnull().sum()} nulos")

        # Aplicar estratégias de tratamento de nulos

print("\n=== APLICANDO ESTRATÉGIAS DE TRATAMENTO ===\n")

# Estratégias definidas:

# 1. Coluna 'data' - Remover linhas com data inválida (crítica para análise temporal)
linhas_antes = len(df_clean)
df_clean = df_clean.dropna(subset=['data'])
print(f"1. Data: removidas {linhas_antes - len(df_clean)} linhas com data inválida")

# 2. Coluna 'receita' - Remover linhas com receita nula (crítica para KPIs)
linhas_antes = len(df_clean)
df_clean = df_clean.dropna(subset=['receita'])
print(f"2. Receita: removidas {linhas_antes - len(df_clean)} linhas com receita nula")

# 3. Coluna 'lucro' - Preencher com 0 (pode ser que alguns produtos tenham lucro 0)
if 'lucro' in df_clean.columns:
    nulos_lucro = df_clean['lucro'].isnull().sum()
    df_clean['lucro'] = df_clean['lucro'].fillna(0)
    print(f"3. Lucro: preenchidos {nulos_lucro} valores nulos com 0")

# 4. Coluna 'quantidade' - Preencher com mediana
if 'quantidade' in df_clean.columns:
    nulos_qtd = df_clean['quantidade'].isnull().sum()
    mediana_qtd = df_clean['quantidade'].median()
    df_clean['quantidade'] = df_clean['quantidade'].fillna(mediana_qtd)
    print(f"4. Quantidade: preenchidos {nulos_qtd} valores nulos com mediana ({mediana_qtd:.0f})")

# 5. Colunas categóricas - Preencher com 'Não informado'
colunas_categoricas = ['uf', 'canal_venda', 'categoria', 'segmento']
for col in colunas_categoricas:
    if col in df_clean.columns and df_clean[col].isnull().sum() > 0:
        nulos = df_clean[col].isnull().sum()
        df_clean[col] = df_clean[col].fillna('Não informado')
        print(f"5. {col}: preenchidos {nulos} valores nulos com 'Não informado'")

print("\n=== NULOS APÓS TRATAMENTO ===")
nulos_depois = df_clean.isnull().sum()
print(nulos_depois[nulos_depois > 0] if (nulos_depois > 0).any() else "✅ Nenhum valor nulo remanescente!")
```

## 7. Padronização de categorias

Os slides mostram que categorias duplicadas ou mal escritas podem distorcer rankings e filtros. fileciteturn2file0

### Tarefa
Inspecione e padronize, se necessário:
- `uf`
- `canal`
- `categoria`

### Pense em:
- maiúsculas e minúsculas
- espaços extras
- acentuação / variações
- categorias equivalentes escritas de formas diferentes


```python
# ## 7. Padronização de categorias

print("=== PADRONIZAÇÃO DE CATEGORIAS ===\n")

# Função para padronizar strings
def padronizar_texto(texto):
    if pd.isna(texto) or texto == 'Não informado':
        return texto
    # Remover espaços extras, converter para maiúsculas, remover acentos
    texto = str(texto).strip().upper()
    # Remover acentos (simplificado)
    acentos = {'Á': 'A', 'À': 'A', 'Ã': 'A', 'Â': 'A', 'É': 'E', 'Ê': 'E', 'Í': 'I', 'Ó': 'O', 'Ô': 'O', 'Õ': 'O', 'Ú': 'U', 'Ç': 'C'}
    for acento, sem_acento in acentos.items():
        texto = texto.replace(acento, sem_acento)
    return texto

# 1. Padronizar UF
if 'uf' in df_clean.columns:
    print("1. Padronizando coluna 'uf':")
    print(f"   Valores únicos antes: {sorted(df_clean['uf'].unique())}")
    df_clean['uf'] = df_clean['uf'].apply(padronizar_texto)
    print(f"   Valores únicos depois: {sorted(df_clean['uf'].unique())}")

# 2. Padronizar canal_venda
if 'canal_venda' in df_clean.columns:
    print("\n2. Padronizando coluna 'canal_venda':")
    print(f"   Valores únicos antes: {sorted(df_clean['canal_venda'].unique())}")
    df_clean['canal_venda'] = df_clean['canal_venda'].apply(padronizar_texto)
    print(f"   Valores únicos depois: {sorted(df_clean['canal_venda'].unique())}")

# 3. Padronizar categoria
if 'categoria' in df_clean.columns:
    print("\n3. Padronizando coluna 'categoria':")
    print(f"   Valores únicos antes: {sorted(df_clean['categoria'].unique())}")
    df_clean['categoria'] = df_clean['categoria'].apply(padronizar_texto)
    print(f"   Valores únicos depois: {sorted(df_clean['categoria'].unique())}")

# 4. Padronizar segmento
if 'segmento' in df_clean.columns:
    print("\n4. Padronizando coluna 'segmento':")
    print(f"   Valores únicos antes: {sorted(df_clean['segmento'].unique())}")
    df_clean['segmento'] = df_clean['segmento'].apply(padronizar_texto)
    print(f"   Valores únicos depois: {sorted(df_clean['segmento'].unique())}")

print("\n✅ Padronização concluída!")
```

## 8. Subindo de nível — Criação de variáveis derivadas

Depois da limpeza, é hora de enriquecer a base com variáveis úteis para análise. fileciteturn2file0

### Tarefa
Crie, se possível:
- `ano`
- `mes`
- `ano_mes`
- `margem_lucro`

### Atenção
A criação de `margem_lucro` exige cuidado com divisões por zero e possíveis valores infinitos.


```python
# ## 8. Subindo de nível — Criação de variáveis derivadas

print("=== CRIAÇÃO DE VARIÁVEIS DERIVADAS ===\n")

# 1. Extrair ano e mês da data
if 'data' in df_clean.columns and pd.api.types.is_datetime64_any_dtype(df_clean['data']):
    df_clean['ano'] = df_clean['data'].dt.year
    df_clean['mes'] = df_clean['data'].dt.month
    df_clean['ano_mes'] = df_clean['data'].dt.strftime('%Y-%m')
    print("✅ Criadas variáveis: ano, mes, ano_mes")
    print(f"   Período: {df_clean['ano'].min()} a {df_clean['ano'].max()}")
    print(f"   Meses disponíveis: {sorted(df_clean['mes'].unique())}")

# 2. Criar margem de lucro (lucro / receita)
if 'lucro' in df_clean.columns and 'receita' in df_clean.columns:
    # Calcular margem, tratando divisão por zero
    # Usar np.where para evitar divisão por zero e valores infinitos
    df_clean['margem_lucro'] = np.where(
        df_clean['receita'] > 0,
        (df_clean['lucro'] / df_clean['receita']) * 100,
        0
    )
    print("\n✅ Criada variável: margem_lucro (%)")
    print(f"   Margem média: {df_clean['margem_lucro'].mean():.2f}%")
    print(f"   Margem mínima: {df_clean['margem_lucro'].min():.2f}%")
    print(f"   Margem máxima: {df_clean['margem_lucro'].max():.2f}%")

    # Verificar se há valores infinitos
    infinitos = np.isinf(df_clean['margem_lucro']).sum()
    if infinitos > 0:
        print(f"   ⚠️ Atenção: {infinitos} valores infinitos encontrados")
        df_clean['margem_lucro'] = df_clean['margem_lucro'].replace([np.inf, -np.inf], np.nan)
        print(f"   Valores infinitos substituídos por NaN")
```
